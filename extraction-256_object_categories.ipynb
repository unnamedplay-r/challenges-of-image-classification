# Image Extraction / Storage

```python
top_path  = './data/256_ObjectCategories'
img1_path = './data/256_ObjectCategories/001.ak47/001_0001.jpg'
img2_path = './data/256_ObjectCategories/001.ak47/001_0002.jpg'
```

```python
directories = os.listdir(top_path)[1:]
directories
```

```python
images = []
for directory in directories:
    pwd = top_path + '/' + directory
    for file in os.listdir(pwd):
        path = pwd + '/' + file
#         print(img)
        images.append((path,directory,file))
```

```python
columns = ['img_path', 'category', 'file_name']
df = pd.DataFrame(images, columns = columns)
```

```python
df['category'] = df.category.str.replace(r'^\d+\.', '')
df['category'] = df.category.str.replace(r'-101$', '')
```

```python
for v in df['category'].value_counts().sort_values(ascending=False).index:
    print(v)
```
